```python
import pandas as pd
import matplotlib.pyplot as plt

a = pd.read_csv("C:\\Users\\marya\\Downloads\\light_curve\\detections.csv")
```

```python
a
```

```python
red_points = a['fid']==2
plt.errorbar(a['mjd'][red_points]-a['mjd'][0], a['magpsf'][red_points], a['sigmapsf'][red_points], c='red', fmt='o')
green_points = a['fid']==1
plt.errorbar(a['mjd'][green_points]-a['mjd'][0], a['magpsf'][green_points], a['sigmapsf'][green_points], c='green', fmt='o')
plt.xlabel("Days")
plt.ylabel("Magnitude")
plt.gca().invert_yaxis()
```
