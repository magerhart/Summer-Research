```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# events with red filter 
plt.figure(dpi=150)
ax1 = plt.subplot(2,1,1)
red_trans_peak = [20.054, 19.854, 19.368]
red_transhost_mag = [20.585, 21.178, 19.75]
ax1.scatter([red_trans_peak], [red_transhost_mag], c='red', label="transient")
red_cv_peak = [17.796, 17.976, 16.572, 17.971, 16.419, 15.779]
red_cvhost_mag = [17.428, 21.963, 22.382, 20.939, 23.256, 18.072]
ax1.scatter([red_cv_peak], [red_cvhost_mag], c='#99250C', marker='v', label="CV")
red_cv_peak_undetermined_mag = [18.792, 15.741, 16.143, 14.882]
red_cv_undetermined_mag = [24, 24, 24, 24]
red_error =  [2, 2, 2, 2]
ax1.errorbar(red_cv_peak_undetermined_mag, red_cv_undetermined_mag, red_error, c= '#99250C', marker='|', lw=0, elinewidth=1.0)
ax1.legend()
plt.subplot(2,1,1)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
plt.title('Host Brightness vs Event Peak Brightness')

# adding line
xplot = np.linspace(0, 30, 10000)
yplot = xplot
ax1.plot(xplot, yplot, c='black', ls='--', lw=0.5)

# events with green filter
ax2 = plt.subplot(2,1,2)
green_trans_peak = [19.544, 19.295, 18.846]
green_transhost_mag = [20.853, 24.347, 23.281]
ax2.scatter([green_trans_peak], [green_transhost_mag], c='#1DE71D', label="transient")
green_cv_peak = [17.516, 17.693, 16.397, 18.043, 16.119, 16.128]
green_cvhost_mag = [19.49, 23.018, 22.469, 20.527, 23.267, 18.295]
ax2.scatter([green_cv_peak], [green_cvhost_mag], c='#146A14', marker='v', label="CV")
green_cv_peak_undetermined_mag = [18.232, 14.852, 16.224, 14.203]
green_cv_undetermined_mag = [24, 24, 24, 24]
green_error = [2, 2, 2, 2]
ax2.errorbar(green_cv_peak_undetermined_mag, green_cv_undetermined_mag, green_error, c='#146A14', marker='|', lw=0, elinewidth=1.0)
ax2.legend()
plt.subplot(2,1,2)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

# adding line
xplot = np.linspace(0, 30, 10000)
yplot = xplot
ax2.plot(xplot, yplot, c='black', ls='--', lw=0.5)

# plot formatting
ax1.get_shared_x_axes().join(ax1,ax2)
ax1.get_shared_y_axes().join(ax1,ax2)
ax1.set_xlim(20.5, 14)
ax1.set_ylim(26, 14.5)

# adding text
ax1.text(20, 20.7, '18abv')
ax1.text(19.75, 21.95, '20aci')
ax1.text(19.25, 20.25, '20abf')

ax2.text(19.5, 21.45, '18abv')
ax2.text(19.25, 24.95, '20aci')
ax2.text(18.8, 23.68, '20abf')

plt.xlabel("Peak Brightness of Event")
plt.ylabel("Average Brightness of Host Galaxy")
```
