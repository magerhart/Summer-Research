```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
transient_data = pd.read_csv('C:\\Users\\marya\\OneDrive\\Desktop\\Light Curves\\Data_collected_csvversion - Sheet1.csv')
red_time = transient_data['red days spend btw 1/4 and 3/4 mag']
green_time = transient_data['green days spend btw 1/4 and 3/4 mag']
magpeak_red = transient_data['true peak mag in red']
magpeak_green = transient_data['true peak mag in green']
```

```python
# plot of points with red filter
plt.figure(dpi=150)
ax1 = plt.subplot(2,1,1)
red_time_trans = red_time[0:4]
red_time_trans = red_time_trans[[True, True, False, True]]
red_mag_trans = magpeak_red[0:4]
red_mag_trans = red_mag_trans[[True, True, False, True]]
ax1.scatter([red_time_trans], [red_mag_trans], c='red', label="FBOT")
red_time_CV = red_time[5:25]
red_mag_CV = magpeak_red[5:25]
ax1.scatter([red_time_CV], [red_mag_CV], marker='v', c='#99250C', label='CV')

# plot formatting
plt.title('Time spent between 25% and 75% magnitude vs peak magnitude for CVs and FBOTs')
plt.xlabel("Time spent between 25% and 75% magnitude of peak magnitude")
plt.ylabel(f'peak magnitude \n Red Filter')
ax1.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

# plots of points with green filter
plt.figure(dpi=150)
ax2 = plt.subplot(2,1,2)
green_time_trans = green_time[0:4]
green_mag_trans = magpeak_green[0:4]
ax2.scatter([green_time_trans], [green_mag_trans], c='#1DE71D', label="FBOT")
green_time_CV = green_time[5:25]
green_mag_CV = magpeak_green[5:25]
ax2.scatter([green_time_CV], [green_mag_CV], marker='v', c='#146A14', label='CV')

# plot formatting

ax1.get_shared_x_axes().join(ax1,ax2)
ax1.get_shared_y_axes().join(ax1,ax2)
plt.xlabel("Time spent between 25% and 75% magnitude of peak magnitude")
plt.ylabel(f'Fade Time in days \n Green Filter')

ax2.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.subplot(2,1,2)
plt.gca().invert_yaxis()



# Adding text 

ax2.text(3.705, 19.53, '18abv')
ax2.text(4.0653, 14.2, '18abc')
```
