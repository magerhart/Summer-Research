```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
transient_data = pd.read_csv('C:\\Users\\marya\\OneDrive\\Desktop\\Light Curves\\Data_collected_csvversion - Sheet1.csv')
red_fade = transient_data['fade in mag over 3 days in red']
green_fade = transient_data['fade in mag over 3 days in green']
magpeak_red = transient_data['true peak mag in red']
magpeak_green = transient_data['true peak mag in green']
```

```python
# plot of points with red filter
plt.figure(dpi=150)
ax1 = plt.subplot(2,1,1)
red_fade_trans = red_fade[0:4]
red_mag_trans = magpeak_red[0:4]
ax1.scatter([red_fade_trans], [red_mag_trans], c='red', label="FBOT")
red_fade_CV = red_fade[5:25]
red_mag_CV = magpeak_red[5:25]
ax1.scatter([red_fade_CV], [red_mag_CV], marker='v', c='#99250C', label='CV')

# plot formatting
plt.title('Fade Time vs peak magnitude for CVs and FBOTs')
plt.xlabel("fade in magnitude in 3 days after peak magnitude")
plt.ylabel(f'peak magnitude \n Red Filter')
ax1.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

# plots of points with green filter
plt.figure(dpi=150)
ax2 = plt.subplot(2,1,2)
green_fade_trans = green_fade[0:4]
green_mag_trans = magpeak_green[0:4]
ax2.scatter([green_fade_trans], [green_mag_trans], c='#1DE71D', label="FBOT")
green_fade_CV = green_fade[5:25]
green_mag_CV = magpeak_green[5:25]
ax2.scatter([green_fade_CV], [green_mag_CV], marker='v', c='#146A14', label='CV')

# plot formatting

ax1.get_shared_x_axes().join(ax1,ax2)
ax1.get_shared_y_axes().join(ax1,ax2)
plt.xlabel("Rise Time in days")
plt.ylabel(f'Fade Time in days \n Green Filter')

ax2.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.subplot(2,1,2)
plt.gca().invert_yaxis()
```
