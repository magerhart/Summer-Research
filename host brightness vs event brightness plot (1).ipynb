```python
# events with red filter 
plt.figure(dpi=150)
ax1 = plt.subplot(2,1,1)
red_trans_peak = [20.054, 14.202, 19.854, 19.368]
red_transhost_mag = [21.564, 19.103, 21.360, 19.798]
ax1.scatter([red_trans_peak], [red_transhost_mag], c='red', label="transient")
red_cv_peak = [18.998, 18.327, 18.143, 15.975, 16.780, 18.230, 16.463, 16.154, 14.815, 16.216]
red_cvhost_mag = [21.957, 20.710, 19.685, 20.395, 21.752, 19.950, 21.490, 20.903, 20.514, 20.31]
ax1.scatter([red_cv_peak], [red_cvhost_mag], c='#99250C', marker='v', label="CV")
ax1.legend()
plt.subplot(2,1,1)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()
plt.title('Host Brightness vs Event Peak Brightness')

# events with green filter
ax2 = plt.subplot(2,1,2)
green_trans_peak = [19.711, 18.204, 19.558, 18.915]
green_transhost_mag = [21.381, 19.606, 21.838, 19.681]
ax2.scatter([green_trans_peak], [green_transhost_mag], c='#1DE71D', label="transient")
green_cv_peak = [18.530, 18.184, 17.935, 15.190, 16.698, 18.417, 16.385, 14.540, 16.760]
green_cvhost_mag = [21.887, 21.412, 20.474, 21.308, 21.171, 20.366, 22.299, 21.220, 20.951]
ax2.scatter([green_cv_peak], [green_cvhost_mag], c='#146A14', marker='v', label="CV")
ax2.legend()
plt.subplot(2,1,2)
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()

# plot formatting
ax1.get_shared_x_axes().join(ax1,ax2)
ax1.get_shared_y_axes().join(ax1,ax2)

plt.xlabel("Peak Brightness of Event")
plt.ylabel("Average Brightness of Host Galaxy")
```
